### Removing columns

```python
import pandas as pd
```

```python
summer=pd.read_csv("summer.csv")
```

```python
summer.head()
```

#### SO, how do you think I should go on if I want to drop the "Sport" column?

##### Well, one way is to just run the .drop() method

```python
summer.drop(columns="Sport")
```

#### The above code ran successfully and managed to drop the "Sport" column. However, there is one big flaw in it. Wanna see what it is?

```python
summer.head()
```

#### Did you notice anything? 

#### Ans: Even though we had dropped the "Sport" column earlier, it again appears if we run the summer df. Why is this happening?

#### The reason is because the .drop() method by default has the inplace parameter set to False

```python
summer.drop(columns=["Sport","Discipline"], inplace=True)
```

```python
summer.head()
```

#### Here we see that the Sport and Discipline columns have been successfully removed

#### 2nd method: Using the parameter "label" 

```python
summer.drop(labels="Event", axis=1)
```

#### Lets check whether here the inplace is by default set to False as well 

```python
summer.head()
```

#### The very fact that we can see the event column still in the df means that the inplace parameter was set to False

```python
# So the solution for that is:

summer.drop("Event", axis=1, inplace=True)
summer.head()
```

#### In the above code, we can also write "columns" instead of 1

#### There is 1 more way to drop a column i.e. by using the keyword "delete"

```python
del summer["Gender"]
```

```python
summer.head()
```

### Removing Rows

```python
summer=pd.read_csv("summer.csv", index_col="Athlete")
summer.head(10)
```

#### Our task is to remove all the rows that have "Alfred Hajos" in it. How shall we do it?



#### The answer to that is there are many ways to do it. Let check them out:

#### 1st method is dropping "x" by using index parameter

```python
summer.drop(index="HAJOS, Alfred")
```

#### We can also pass a list of "x" to the index parameter

```python
summer.drop(index=["HERSCHMANN, Otto", "DRIVAS, Dimitrios"], inplace=True)
```

```python
summer.head(10)
```

#### 2nd method: using label parameter and axis to denote the column of "x"

```python
summer.drop(labels="HAJOS, Alfred", axis=0, inplace=True)
```

```python
summer.head()
```

#### What if you wanted to have the data for just the 1996 olympic games?

#### Now, lets assumer that we do not want to select some rows that meet certain conditions and we wanna drop those rows that meet certain conditions

#### So, lets assumer that we are not interested in 1996 olympic games and we do not like "acquatic" games. In other words, we wanna remove all the rows that in which the Year is either 1996 or the sport is "Aqcquatics"

```python
summer=pd.read_csv("summer.csv")
summer.head()
```

```python
# To solve this problem, first of all, we have to create 2 boolean series

crit1=summer.Year==1996
crit2=summer.Sport=="Aquatics"
```

```python
summer1=summer.loc[~(crit1|crit2)]
```

```python
summer1.head()
```

#### So, now how to check whether there are "1996" in the df?

```python
(summer1.Year==1996).value_counts() # Since there are only Falses, we can conclude that there is no 1996 in the above df
```

#### What about "Aquatics"?

```python
(summer1.Sport=="Aquatics").value_counts()
```
