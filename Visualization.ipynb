```python
import pandas as pd
```

```python
titanic=pd.read_csv("titanic.csv")
titanic.head()
```

```python
import matplotlib.pyplot as plt
```

```python
titanic.plot() # Obviously, when we call the plot() method, only the numerical values shall be plotted
plt.show()

# This data all together packed into 1 is actually not very helpful while plotting
```

#### So, lets separate the different variables

```python
titanic.plot(subplots=True)
plt.show()
```

#### Lets make the figure a bit bigger

```python
titanic.plot(subplots=True, figsize=(12,8)) # For the figsize, you have to determine the width and the height respectively
plt.show()
```

#### Can we have have the x-axis stated in each of the subplot? 

```python
titanic.plot(subplots=True, figsize=(14,12), sharex=False)
plt.show()
```

#### By default, the sharex is True and sharey is False.

#### Can we plot just 1 column instead of entire df?

```python
titanic.age.plot(figsize=(12,8))
plt.show()
```
