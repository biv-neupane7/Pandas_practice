#### Lets customize our plot a little bit:

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
titanic=pd.read_csv("titanic.csv")
titanic.head()
```

#### Can we make the fontsize in x and y axis a bit bigger?

```python
titanic.age.plot(figsize=(12,10), fontsize=20)
plt.show()
```

#### Lets now change the color of the plot

```python
titanic.age.plot(figsize=(12,10), fontsize=13, color="red") # Instead of color and red, you can also write c and "r"
plt.show()
```

#### What if we wanna change the linestyle?

```python
titanic.age.plot(figsize=(12,10), fontsize=13, c="r", linestyle="--")
plt.show()
```

#### We have to look into the documentation to know what are the various kinds of colors and linestyles that the pandas provides. For now, lets back to the original line style:

```python
titanic.age.plot(figsize=(12,10), fontsize=13, c="r", linestyle="-")
plt.show()
```

#### What about the title of the plot? How would I know what the plot is all about?

```python
titanic.age.plot(figsize=(12,8), fontsize= 13, c="r", linestyle="-")
plt.title("Age of the people in titanic", fontsize=15)
plt.show()
```

#### There is one variable in the plot. But, if there were more than 1, can we add a legend in the plot that will help the readers navigate the plot?

```python
titanic.age.plot(figsize=(12,8), fontsize=13, c="r", linestyle="-")
plt.title("Age of the people in titanic", fontsize=15)
plt.legend(loc=0, fontsize=15)
```

#### Have you noticed that we still do not know what the x-axis and y-axis actually represent?

```python
titanic.age.plot(figsize=(12,8), fontsize=13, c="r", linestyle="-")
plt.title("Age of the people in titanic", fontsize=15)
plt.legend(loc=0, fontsize=15)
plt.xlabel("No. of Passengers", fontsize=13)
plt.ylabel("Age", fontsize=13)
plt.show()
```

#### To see the plot a bit more granularly, we can also introduce grids:

```python
titanic.age.plot(figsize=(12,8), fontsize=13, c="r", linestyle="-")
plt.title("Age of the people in titanic", fontsize=15)
plt.legend(loc=0, fontsize=15)
plt.xlabel("No. of Passengers", fontsize=13)
plt.ylabel("Age", fontsize=13)
plt.grid()
plt.show()
```

#### What if we want to change x and yticks?

```python
xtik=[x for x in range(0,901,50)]
ytik=[y for y in range(0,81,5)]
```

```python
titanic.age.plot(figsize=(12,8), fontsize=13, c="r", linestyle="-", xticks=xtik, yticks=ytik)
plt.title("Age of the people in titanic", fontsize=15)
plt.legend(loc=0, fontsize=15)
plt.xlabel("No. of Passengers", fontsize=13)
plt.ylabel("Age", fontsize=13)
plt.grid()
plt.show()
```

#### how to create a histogram?

```python
import pandas as pd
import matplotlib.pyplot as plt
plt.style.use("seaborn")
```

```python
titanic=pd.read_csv("titanic.csv")
```

```python
titanic.head()
```

```python
titanic.age.value_counts()
```

#### Can we sort this age data into various bins?

```python
titanic.age.value_counts(bins=10) # You can check whether its equal weight bins by subtracting the higher range by 
                                  # lower range. Here the difference is 7.958 
```

#### Now, lets potray this above bins in a plot namely, a histogram..!

```python
titanic.age.plot(kind="hist", figsize=(8,12), fontsize=13)
plt.show()
```

#### You might have noticed that in the above code, we did not mention anything about bins. Nevertheless, the data was divided into 10 bins. This is because sorting into 10 bins is by default. However, we can easily change this.

```python
titanic.age.plot(kind="hist", figsize=(12,8), fontsize=13, bins=80)
plt.show()
```

#### This histogram mentioned above can also be created in two other different ways. The first one is:

```python
titanic.age.hist(figsize=(12,8), bins=80, xlabelsize=13, ylabelsize=13)
plt.show()
```

#### The 2nd way is:

```python
plt.figure(figsize=(12,8))
plt.hist(titanic.age.dropna(), bins=80, density=False, cumulative=False)
plt.show()

# Note that this method cannot handle the missing values in the data. So, if your data has any na values, make sure to drop
# them.
```

#### So, as you can see, each method has some variations in it and serves different purposes.

### Scatter plots

```python
import pandas as pd
import matplotlib.pyplot as plt
plt.style.use("seaborn")
```

```python
titanic=pd.read_csv("titanic.csv")
titanic.head()
```

```python
titanic.plot(figsize=(12,8), kind="scatter", x="age", y="fare")
plt.show()
```

#### If I want to change the color, what should I do?

```python
titanic.plot(kind="scatter", figsize=(12,8), x="age", y="fare", c="red")
plt.show()
```

#### Now, what if I wanted to have diamonds instead of dots OR any other way to visualize the data points?

```python
titanic.plot(kind="scatter", figsize=(12,8), x="age", y="fare", c="red", marker="D")
plt.show()



# You can add many variations in the marker section such as "x", "X", "D", etc.

# You can also change the size of the dots by using s="20" or s="50" or whatever number suits you
```

#### How to convert this 2 dimensional plot into a 3 dimensional plot?

```python
# Well if we simply add 1 more variable into the mix, then it might be possible

titanic.plot(kind="scatter", figsize=(12,8), x="age", y="fare", c="survived", marker="x")
plt.show()
```

#### But I cannot see anything in the plot. How can I even start interpreting the results? Make it a bit more readable

```python
titanic.plot(kind="scatter", figsize=(12,8), x="age", y="fare", c="survived", marker="x", colormap="viridis")
plt.show()
```

#### In the above plot, we cannot see all the colors in action because the survived column has only two sets of datpoints i.e. 0 and 1. 

#### Lets try to put another datapoint and check the result

```python
titanic.plot(kind="scatter", figsize=(12,8), x="age", y="fare", c="pclass", marker="x", colormap="viridis")
plt.show()

# Now you can see all the three colors in action
```

#### So from the plot above, a relationship can be observed:

#### The young people tended to buy the 3rd class tickets and the older passengers tend to have 1st class tickets

## Advanced /Statistical plotting with Seaborn

```python
import seaborn as sns
```

#### Make a countplot using seaborn

```python
plt.figure(figsize=(12,8))
sns.countplot(data=titanic, x="sex")
plt.show()
```

#### If we pass the sex column to the y-axis, it will give out a horizontal countplot

```python
plt.figure(figsize=(12,8))
sns.countplot(data=titanic, y="sex")
plt.show()
```

#### Lets further sub-divide the male and female groups by passenger class

```python
plt.figure(figsize=(12,8))
sns.countplot(data=titanic, x="sex", hue="pclass")
plt.show()
```

#### Is there a way to customize this seaborn plot?

#### Yes, by using the set () method, we can perform various operation. Feel free to check out the documentation

```python
plt.figure(figsize=(12,8))
sns.countplot(data=titanic, x="sex", hue="pclass")
sns.set(font_scale =2, palette="viridis") # There are various other options to choose from
plt.show()
```

### Categorical plots with seaborn

```python
plt.figure(figsize=(12,8))
sns.set(font_scale=1.5)
sns.stripplot(data=titanic, x="sex", y="age", hue="pclass")
plt.show()
```

#### In the above plot, althought the data of pclass has been plotted, due to the overalapping nature of the plot, we cannot infer anything meaningful. So, what do we do now?

```python
plt.figure(figsize=(12,8))
sns.set(font_scale=1.5)
sns.stripplot(data=titanic, x="sex", y="age", hue="pclass", jitter=True)
plt.show()
```

#### Alright, this is certainly an improvement from before. But, still the datapoints for different classes among male and female are all mixed up. I wish there was a way to separate the pclasses within each sex

```python
plt.figure(figsize=(12,8))
sns.set(font_scale=1.5)
sns.stripplot(data=titanic, x="sex", y="age", hue="pclass", jitter=True, dodge=True)
plt.show()
```

#### From the above figure, (for both sexes), it looks like that the 1st class passengers are older than the 2nd class and so on.

## Swarmplot

#### Swarmplot: It is exactly like the stripplot except for the fact that we can actually the distribution of the data in each sub segment

```python
plt.figure(figsize=(12,8))
sns.set(font_scale=1.5)
sns.swarmplot(data=titanic, x="sex", y="age", hue="pclass", dodge=True) # Also notice that there is parameter called jitter
plt.show()
```

#### We can infer from the above plot that in the male category, the 3rd class was mainly filled with 20-23 year old guys
